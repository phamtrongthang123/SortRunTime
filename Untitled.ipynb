```python
%matplotlib inline
import pandas as pd
import numpy as np
data1 = pd.read_csv("out_ngaunhien.csv", names=["Sort", "Time", "N"])
```

```python
data1.head(130)
```

```python
n = data1.shape[0]
```

```python
list_sort = [];
for i in range(13):
    list_sort.append(data1['Sort'][i])
list_sort
```

```python
list_N = list(data1['N'].value_counts().index)
list_N = sorted(list_N)
list_N
```

```python
table2 = {
    'Sort type' : list_sort[5:]
}
table2
```

```python
table1 = {
    'Sort type' : list_sort,
}
for i in list_N[:10]:
    table1[i] = []
j =0; k=0; tmp = 0;
for i in data1['Time']:
    if j < 130:
        if tmp < 13:
            tmp = tmp + 1
        else:
            k = k +1;
            tmp = 1;
        table1[list_N[k]].append(i);
    else:
        break;
    j = j +1;
               
        
table1
```

```python
data_ngaunhien1 = pd.DataFrame(table1).set_index('Sort type').T
data_ngaunhien1
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(400,400))
data_ngaunhien1.plot(figsize=(12,4))
plt.show()
```
