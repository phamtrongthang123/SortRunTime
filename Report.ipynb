```
%matplotlib inline
import pandas as pd
import numpy as np
data1 = pd.read_csv("SortRunTime/out_ngaunhien.csv", names=["Sort", "Time", "N"])
data2 = pd.read_csv("SortRunTime/out_dasapxep.csv", names=["Sort", "Time", "N"])
data3 = pd.read_csv("SortRunTime/out_nguocthutu.csv", names=["Sort", "Time", "N"])
data4 = pd.read_csv("SortRunTime/out_gannhusapxep.csv", names=["Sort", "Time", "N"])
data_bosung = pd.read_csv("SortRunTime/out_bosungRadixsort.csv");
```

```
list_bosung = list(data_bosung.set_index('0').T)
list_bosung
```

```

def proc(data1):
  list_sort = [];
  for i in range(13):
      list_sort.append(data1['Sort'][i])


  list_N = list(data1['N'].value_counts().index)
  list_N = sorted(list_N)

  table2 = {
      'Sort type' : list_sort[5:]
  }
  table1 = {
      'Sort type' : list_sort,
  }
  for i in list_N[:10]:
      table1[i] = []
  for i in list_N[10:]:
      table2[i] = []
  j =0; k=0; tmp = 0;
  for i in data1['Time']:
      if j < 130:
          if tmp < 13:
              tmp = tmp + 1
          else:
              k = k +1;
              tmp = 1;
          table1[list_N[k]].append(i);
      else:
          if tmp < 8:
              tmp = tmp + 1
          else:
              k = k +1;
              tmp = 1;
          table2[list_N[k]].append(i);
      j = j +1;  
  return table1, table2;
```

```
table1, table2 = proc(data1)
data_ngaunhien1 = pd.DataFrame(table1).set_index('Sort type').T
data_ngaunhien2 = pd.DataFrame(table2).set_index('Sort type').T
table1, table2 = proc(data2)
data_dasapxep1= pd.DataFrame(table1).set_index('Sort type').T
data_dasapxep2= pd.DataFrame(table2).set_index('Sort type').T
table1, table2 = proc(data3)
data_nguocthutu1= pd.DataFrame(table1).set_index('Sort type').T
data_nguocthutu2= pd.DataFrame(table2).set_index('Sort type').T
table1, table2 = proc(data4)
data_gannhusapxep1= pd.DataFrame(table1).set_index('Sort type').T
data_gannhusapxep2= pd.DataFrame(table2).set_index('Sort type').T
```

```
n = [1,2,4,6,8,10,12,14,16,20]
k = 10000
j = 0;
for i in range(10):
  data_ngaunhien1['Radix sort'][n[i] * k] = list_bosung[j];
  data_dasapxep1['Radix sort'][n[i] * k] = list_bosung[j+1];
  data_nguocthutu1['Radix sort'][n[i] * k] = list_bosung[j+2];
  data_gannhusapxep1['Radix sort'][n[i] * k] = list_bosung[j+3];
  j = j + 4
k = 1000000
j = j -1;
for i in range(10,20):
  data_ngaunhien2['Radix sort'][n[i-10] * k] = list_bosung[j];
  data_dasapxep2['Radix sort'][n[i-10] * k] = list_bosung[j+1];
  data_nguocthutu2['Radix sort'][n[i-10] * k] = list_bosung[j+2];
  data_gannhusapxep2['Radix sort'][n[i-10] * k] = list_bosung[j+3];
  j = j +4;
```
